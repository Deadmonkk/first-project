```python
import numpy as np
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt 
```

```python
import pandas as pd
```

```python
df= pd.read_csv('/Users/sulu/Downloads/VideoGamesSales.csv')
#df= df[df.duplicated()] to check number of duplicates.
df= df.drop_duplicates()

```

```python
df.info()
```

```python
df['Region'] = df['Region'].fillna('North')
```

```python
df.info()
```

```python
df['Publisher'] = df['Publisher'].fillna('Nintendo')
```

```python
df.info()
```

```python
#for the bottom 100 values
df.tail(100)
```

```python
df
```

```python
# Replacing the $ sign and making the data consistent. 
df['NA_Sales'] = df['NA_Sales'].replace('[$]','', regex=True)
```

```python
df
```

```python
#Replacing the national sales empty values with the average value.

df['NA_Sales'] = pd.to_numeric(df['NA_Sales'], errors ='coerce')
average_sales = df['NA_Sales'].mean()
average_sales = int(average_sales)
df['NA_Sales'] =df['NA_Sales'].fillna(average_sales)
```

```python
#Making the country column consistent.
df['Country'] = df['Country'].replace({'United States':'USA'})
#df = df[df['Country']=='USA']
df.shape[0]
```

```python
#Renaming Columns.
df= df.rename(columns={'NA_Sales': 'National Sales', 'Global_Sales':'Global Sales', 'Global_Profit':'Global Profit', 'NA_Profit': 'National Profit'})
```

```python
#Working on Outliers.
sales_cap = df['National Sales'].quantile(0.95)
df['National Sales'] = np.where(df['National Sales'] > sales_cap, sales_cap, df['National Sales'])
df
```

```python
#Data Visualization.

#Bar-graph
national_sales = df.groupby(['Region', 'Country'])['National Sales'].sum().reset_index().sort_values(by= 'National Sales', ascending = False)
plt.figure(figsize= (16,9))
sns.barplot(data= national_sales, x = 'Region', y = 'National Sales', hue = 'Country')
plt.title('National Sales by Country and Region', fontweight= 'bold')
plt.xlabel('Region', fontweight= 'bold')
plt.ylabel('National Sales', fontweight= 'bold')
plt.show()

```

```python
#BOX Plot:

plt.figure(figsize=(16,9))
sns.boxplot(x='Country', y='National Sales', data =df, showmeans=True, hue='Genre', meanprops= {"marker":"o", "markerfacecolor":"white", "markersize":"6", "markeredgecolor":"green"}, showfliers=False)
plt.title('National Sales by Country', fontweight= 'bold')
plt.xlabel('Country', fontweight= 'bold')
plt.ylabel('National Sales', fontweight= 'bold')
plt.show()
```

```python
#Pie Chart.
sales = df.groupby(['Country'])[['National Sales', 'Global Sales']].sum().reset_index().sort_values(by= 'Country', ascending = False)
Country_is = sales['Country']
National_Sales = sales['National Sales']
Global_Sales = sales['Global Sales']

fig, axs = plt.subplots(1, 2, figsize=(16, 9))

axs[0].pie(National_Sales, labels= Country_is, autopct='%1.1f%%', startangle=90)
axs[0].set_title('National Sales by Country')

axs[1].pie(Global_Sales, labels= Country_is, autopct='%1.1f%%', startangle=90)
axs[1].set_title('Global Sales by Country')

plt.show()
```

```python
#Line Chart.

plt.figure(figsize=(16, 9))

#NATIONAL
plt.plot(df['Year'], df['National Sales'], marker='o', linestyle='-', color='b', label='National Sales')

#GLOBAL SALES
plt.plot(df['Year'], df['Global Sales'], marker='x', linestyle='--', color='g', label='Global Sales')


plt.title('National and Global Sales Over Years')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.legend()

plt.grid(True)
plt.show()
```
